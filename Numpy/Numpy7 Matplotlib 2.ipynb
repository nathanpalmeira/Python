```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sal1= np.array([100,200,400,300])
sal2= np.array([50,300,500,450])
```

```python
plt.plot(sal1, c="Black", ls='--', marker='s', ms=8, label="Sal치rio 1") 
plt.plot(sal2, c="Red", ls='--', marker='s', ms=8, label="Sal치rio 2") 
plt.show()
```

```python
plt.plot(sal1, c="Black", ls='--', marker='s', ms=8, label="Sal치rio 1") 
plt.plot(sal2, c="Red", ls='--', marker='^', ms=8, label="Sal치rio 2") 
plt.legend()
plt.show()
```

```python
plt.plot(sal1, c="Black", ls='--', marker='s', ms=8, label="Sal치rio 1") 
plt.plot(sal2, c="Red", ls='--', marker='^', ms=8, label="Sal치rio 2") 
plt.legend(loc='lower right')#alterando localiza칞칚o da legenda
plt.show()
```

```python
sal3 =[200,150,300,700]

plt.plot(sal1, c="Black", ls='--', marker='s', ms=8, label="Sal치rio 1") 
plt.plot(sal2, c="Red", ls='--', marker='^', ms=8, label="Sal치rio 2") 
plt.plot(sal3, c="Green", ls='--', marker='o', ms=8, label="Sal치rio 3") 
plt.legend(loc='lower right')#alterando localiza칞칚o da legenda
plt.show()
```
